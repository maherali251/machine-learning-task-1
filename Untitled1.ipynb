```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in Iris dataset
df = sns.load_dataset('iris')

# Show the shape, column names, and first few rows
print("Shape of dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())

# Show general info about the dataset
print("\nDataset Info:")
print(df.info())

# Show basic statistics for each numeric column
print("\nSummary Statistics:")
print(df.describe())

# Create scatter plots to see relationships between features, colored by species
sns.pairplot(df, hue='species')
plt.suptitle("Scatter Plots of Features by Species", y=1.02)
plt.show()

# Create histograms to check the distribution of each feature
df.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.tight_layout()
plt.show()

# Create box plots to detect outliers for each feature by species
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns[:-1], 1):  # Skip the 'species' column
    plt.subplot(2, 2, i)
    sns.boxplot(x='species', y=column, data=df)
    plt.title(f'Box Plot of {column}')
    plt.xlabel('')
    plt.ylabel('')
plt.suptitle("Box Plots of Features by Species", fontsize=16, y=1.02)
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
